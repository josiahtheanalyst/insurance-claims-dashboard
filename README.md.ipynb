# üìä Finalcial Insurance Claims' & Risk' Interactive Dashboard

## Overview
This project presents an **interactive insurance dashboard** built using **Python (Plotly & Pandas)**. It provides a comprehensive overview of customer risk, loan, and claim data, enabling stakeholders to make data-driven decisions efficiently. The dashboard is designed for **financial analysts, insurance managers, and risk assessment teams**. It's an interactive financial insurance dashboard analyzing customer credit risk, claims, and loan approval recommendations.

## üîß Tools Used
- Python (Pandas, Scikit-learn, Seaborn, Plotly)
- Google Colab
- Machine Learning: Random Forest

## üìà Key Features

- **KPI Cards**: Quick overview of key metrics:
  - Average Credit Score
  - Default Rate (%)
  - Claim Approval Rate (%)
  - Loss Ratio (%)

- **Risk Analysis**
  - Risk distribution by category (interactive donut chart)
  - Risk segmentation by region (heatmap)
  - Default rate by risk category

- **Loan & Credit Insights**
  - Loan-to-income ratio trends by region
  - Loan amount by employment status
  - Credit Score vs Default Rate (scatter with trendline)

- **Claims Analysis**
  - Claims filed per month (interactive line chart)
  - Claim outcomes and status counts (interactive bar/donut charts)

- **Interactive visualizations (Plotly)**

- **Feature importance via Random Forest**

- **Loan approval recommendations by credit score**

## Dataset
The dashboard uses an internal dataset (`df_inner`) containing anonymized customer insurance data with the following columns:

- `customer_id`
- `income`
- `loan_amount`
- `credit_score`
- `claim_amount`
- `claim_status`
- `default_history`
- `risk_category`
- `region`
- `employment_status`
- `incident_date`
